find names with unique intervals, 
non-unique intervals merge the min and max interval ages into a different dataframe, 
keep reference #s accepted names and intervals with raw min max data 

```python
import pandas as pd
```

```python
data=pd.read_csv("../Documents/pbdb_nolit.csv")
data
```

```python
fossil=data.dropna(axis=0, subset=["max_ma", "min_ma"])
fossil
```

```python
uniq_mins = fossil.groupby('accepted_name')["min_ma"].unique().to_frame()
uniq_mins
```

```python
together = pd.concat([fossil, uniq_mins], axis=0, sort=True)
together
```

```python
fossil.min_ma.unique()
```

```python
uniq_maxs = fossil.groupby('accepted_name')["max_ma"].unique().to_frame()
uniq_maxs
```

```python
again = pd.concat([together, uniq_maxs], axis=0, sort=True)
again
```

```python
fossils=again.dropna(axis=0, subset=["max_ma", "min_ma"])
fossils
```
