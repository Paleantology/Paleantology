```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
```

```python
no_corr = pd.read_csv("mk_nocorr.log", delimiter="\t")
```

```python
mk_plain = pd.read_csv("mk.log", delimiter="\t")
```

```python
mk_disc = pd.read_csv("mk_discretized.log", delimiter="\t")
```

```python
mk_cat = pd.read_csv("binary_cat.log", delimiter="\t")
```

```python
no_corr['Model'] = "No Correction"
mk_plain['Model'] = "Mk"
mk_disc['Model'] = "Discretized Beta"
mk_cat['Model'] = "SHDM"
```

```python
no_corr
```

```python
result = pd.concat([no_corr, mk_plain, mk_disc,mk_cat])
```

```python
slim = result[['Likelihood', 'Model']]
```

```python
grouby = slim.groupby('Model')
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.set_facecolor("white")
fig.set_xlabel = "Likelihood"
ax.facecolor=("white")
ax.set_xlabel("Likelihood")
bp = slim.groupby('Model').plot(kind='kde', ax=ax, xlim=[-250, -400])
ax.legend(['Discrete Beta','Corrected','No Correction','SHDM'], facecolor="white")
```

```python
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(18,6))
for name, group in slim.groupby('Model'):
    group.plot(x=group['Model'], ax=ax, xlim=[-250, -400])

plt.show()
```
