I just have some brief code and output for comparing the corrected model in MB and RB. It looks like they find almost
identical L, but different TL.



```python
import pandas as pd
```

```python
%matplotlib notebook

import matplotlib.pyplot as plt
```

```python
#treelength plots
mb_tl = pd.read_csv("mb_tl.csv", index_col=0, header=None)
rb_tl = pd.read_csv("rb_tl.csv", index_col=0, header=None)
mb_tl_uncorr = pd.read_csv("mb_TL_uncorr.csv", index_col=0, header=None)
rb_tl_uncorr = pd.read_csv("rb_TL_uncorr.csv", index_col=0, header=None)
mb_tl['Software'] = "MrBayes"
rb_tl['Software'] = "RevBayes"
mb_tl_uncorr['Software'] = "MrBayes - No correction"
rb_tl_uncorr['Software'] = "RevBayes - No correction"
```

```python
#treelength plots

mb_tl.columns = ["TL", "Software"]
rb_tl.columns = ["TL", "Software"]
mb_tl_uncorr.columns = ["TL", "Software"]
rb_tl_uncorr.columns = ["TL", "Software"]


mb_tl
```

```python
#treelength plots

megadf = pd.concat([mb_tl, rb_tl,mb_tl_uncorr, rb_tl_uncorr])
megadf 
```

```python
#treelength plots

fig, ax = plt.subplots(figsize=(8,6))
megadf.groupby("Software").plot(kind="kde", ax=ax)
ax.legend(['MrBayes', 'MrBayes - No Correction','RevBayes', 'RevBayes - No Correction' ])
plt.xlabel("Tree Length")
```

```python
#L plots
mb_l = pd.read_csv("mb_L.csv", index_col=0, header=None)
rb_l = pd.read_csv("rb_L.csv", index_col=0, header=None)
mb_l_u = pd.read_csv("mb_L_uncorr.csv", index_col=0, header=None)
rb_l_u = pd.read_csv("rb_L_uncorr.csv", index_col=0, header=None)
mb_l['Software'] = "MrBayes"
rb_l['Software'] = "RevBayes"
mb_l_u['Software'] = "MrBayes - No correction"
rb_l_u['Software'] = "RevBayes - No correction"
```

```python
#treelength plots

mb_l.columns = ["L", "Software"]
rb_l.columns = ["L", "Software"]
mb_l_u.columns = ["L", "Software"]
rb_l_u.columns = ["L", "Software"]
mb_l
```

```python
#treelength plots

megaldf = pd.concat([mb_l, rb_l,mb_l_u, rb_l_u,])
megaldf 
```

```python
#treelength plots

fig, ax = plt.subplots(figsize=(8,6))
megaldf.groupby("Software").plot(kind="kde", ax=ax)
ax.legend(['MrBayes', 'MrBayes - No Correction','RevBayes', 'RevBayes - No Correction' ])
plt.xlabel("Likelihood Score")
plt.xlim([-15000, -8000])
```
